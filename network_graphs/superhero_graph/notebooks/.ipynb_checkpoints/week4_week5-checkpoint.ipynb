Some of the character names & associated wiki links seem odd, the char name does not seem to match the url. This is for example the case for the char "Vectron" that has the wiki url "Elseworld's Finest: Supergirl & Batgirl#Secondary characters and cameos", and "Van'n Intraktus, Magaan" that has the wikilink "Green Lantern Corps#First attempt of rebuilding the Green Lantern Corps". This seem to arise from the situation when a wiki page contains a char links for certain character, that does not have their own wiki page, but instead is directed to another page that mention this char. Many of these incidens seem to arise from the page https://en.wikipedia.org/wiki/List_of_DC_Comics_characters:_V  where we have a list of links from character names, to wikilinks that are not specfically describing the character.

If we select what our csv believe is a character page, that in some cases may lead to a more general page, and we then download the content of this page and find wikilinks, and then create connections based on what we interpreted as a specific character based on that, we will end up creating a lot of connections to eg. 'Vectron', although the links was in fact just leading to the Green Lantern Corpse. 

As an example the above problem has linked the char Val Colby to the 'Lex Luthor#Relationships and family' wiki, and the consequence of the above is that the network think that Val Colby with the wikilink 'Lex Luthor#Relationships and family' has 32 connections, whereas in fact it has 0.

Problem is that the vissible wiki link (2nd part) is not nescesarily the name of the char, fx. "Doctor Fate|Fate" "Doctor Fate" and "Doctor Fate|Nabu" all appear as internal links, linking to the same wiki page. Hence the vissible link is sometimes showed as just Fate for short, and for another completely different reference, Nabu, links to the Doctor Fate page. For that reason it is better to stick to the WikiUrls that are unique. 

The most robust method is therefore to discard characters that does not seem to have their own specific page, before creating links from their wiki. Another method could be to create the character if mentioned, but not create any links based on the wiki-page if the wikipage does not seem to represent that specific character, but that would simply lead to isolated nodes as a character with no "legitimate" character wiki-page, would have no inbound or outbound leads. 

```python
import json
import urllib.request

import os.path

baseurl = "https://en.wikipedia.org/w/api.php?"
action = "action=query"
prop = "prop=revisions&rvprop=content" 
dataformat ="format=json"


char_file_list_dc = []  
char_file_list_marvel = []


for df in [dfdc_raw,dfmarvel_raw]:
    for character in df['WikiLink']:
        character_url = character.replace(' ','_') #replacing white spaces with _ for the query, not handled by Urllib quote
        character_url = urllib.parse.quote(character_url) #Use URL parser to convert special chars to URL friendly format
        character_file = character_url.replace('/','~') #some wiki URLS contains / that can not part of files saved on OS, but has to be part of the wikiUrl for the page to work
        if df.equals(dfdc_raw):
            char_file_list_dc.append(character_file+'.json')
        else:
            char_file_list_marvel.append(character_file+'.json')
            
        if os.path.isfile(f"{character_file}.json") == False: 
        #create query for each character
            title = f'titles={character_url}'
            query = "{}{}&{}&{}&{}".format(baseurl, action, prop, title, dataformat)
            #parse quote to format special characters 
            with urllib.request.urlopen(query) as wikiresponse:
                wikitext = wikiresponse.read().decode('utf-8')
            wikijson = json.loads(wikitext)
            with open(f"{character_file}.json", "w") as outfile:
                json.dump(wikijson, outfile)

dfdc_raw['file'] = char_file_list_dc
dfmarvel_raw['file'] = char_file_list_marvel
    
```

```python
#Working

dfdc_clean = dfdc_raw.copy()
dfmarvel_clean = dfmarvel_raw.copy()

#Intention to delete rows where character does not have their own WikiLink
#for example; character: Yellow Lantern, WikiLink: Hal Jordan#Other versions 653
#1st idea, simply sort out rows that does not have an exact name match from char to url
#However, also edge cases like char: Man-Spider url: Spider-Man, or Mister Fish' & 'Mr. Fish
#Instead i search if ANY word from char, matches the URL. This may create some false positives


dim1_dc =dfdc_clean.shape
dim1_marvel =dfmarvel_clean.shape

#To check what chars are removed
soft_filter = []

for df in [dfdc_clean, dfmarvel_clean]:
    for character, wikilink, index in zip(df.CharacterName, df.WikiLink, df.index):
        #Turned out there was still majorty false positives,
        #replace - with whitespace, and split words at whitespaces
        if all(word not in wikilink for word in character.replace('-',' ').split(" ")) and "#" in wikilink:
            #print(character, wikilink)
            soft_filter.append((character, wikilink))
            df.drop(index, inplace=True)

dim2_dc = dfdc_clean.shape
dim2_marvel = dfmarvel_clean.shape
print(f'A total of {(dim1_dc[0]-dim2_dc[0])} DC chars was deleted')

print(f'A total of {(dim1_marvel[0]-dim2_marvel[0])} Marvel chars was deleted')
```

For redirect> Discovered that some chars that have multiple page links, may have a re-direct in one of their sections like "Wolverine DC", "Wolverine Movie #redirect", and was wrongly discarded. I then checked the length of all discarded rows, and in generall the size of these documents are in the range of 700-1100 bytes , compared to red tornado at 63927 or wolverine at 219940

```python
#Check all the downloaded for redirects, the documents contain the exact phrase #REDIRECT
#Remove files with redirect in the documenets

dim1_dc =dfdc_clean.shape
dim1_marvel =dfmarvel_clean.shape

for df in [dfdc_clean, dfmarvel_clean]:
    for index, file in df['file'].iteritems():
        f = open(file, 'r')
        redirect = re.findall(r'#redirect', f.read(), re.IGNORECASE)
        if redirect:           
            #Empircally found reasonable threshold at 2000 bytes
            if os.path.getsize("./"+file) < 2000:
                print(f'Redirect found! in {file}, dropped row')
                df.drop(index, inplace=True)
                if os.path.exists(file) == True:
                    try: 
                        os.remove("./"+file)
                        print(f'removed {file}')
                    except:
                        print("Error while deleting file ", file)
                        
                    
dim2_dc = dfdc_clean.shape
dim2_marvel = dfmarvel_clean.shape
print(f'A total of {(dim1_dc[0]-dim2_dc[0])} DC chars was deleted')

print(f'A total of {(dim1_marvel[0]-dim2_marvel[0])} Marvel chars was deleted')
```

#### Finding outgoing wikilinks (and other false positive "links"

```python
outlink_list = []
textlen_list = []

for wikilink,file in zip(df_superhero.WikiLink, df_superhero.file):

    #For some reson the open(file) closes automatically after a single operation,
    # hence I need to open the file twice seperately

    with open(file) as f:
        words = re.findall(r'(\w+)',f.read())
        #Set the node attribute "words" to len of total words 
        textlen_list.append(len(words))

    with open(file) as f:
        outlink = re.findall(r'\[\[(.+?)\]\]', f.read(), re.IGNORECASE)
        outlink_list.append(outlink)
        #This regex has the problem of returning wikilinks potentially as [Thor (Marvel Comics)|Thor] (not split)
        
df_superhero['WikiOutLinks'] = outlink_list
df_superhero['TextLen'] = textlen_list
```

```python
import networkx as nx

#WARNIING: Takes a few minutes to run...
#def superhero_network(superdf):
G_superhero = nx.DiGraph()
G_superhero.clear()
for fromCharWikiLink, fromUniverse, fromTextLen,outLinkList in zip(df_superhero.WikiLink,df_superhero.universe, df_superhero.TextLen,df_superhero.WikiOutLinks):
    G_superhero.add_node(fromCharWikiLink[0], universe=fromUniverse, textlen=fromTextLen)
    for outlink in outLinkList: 
        for toCharWikiLinkList in df_superhero['WikiLink']: #List of wikilinks, to iterate through to see if outlink exist as a wikilink in our df.
            for toCharLink in toCharWikiLinkList:
                if outlink.split("|")[0] == toCharLink: #From the Regex outlinks has to be split to take the 'wikilink' part of the links that are twofold
                    G_superhero.add_edge(fromCharWikiLink[0], toCharWikiLinkList[0])
```

### Task: Let's generate the Random (Erdös-Renyi) network which has a degree distribution that matches the SuperHero network

Knowing the values of $N$ and number of edges we can calculate $k$, and then $p$, the probability of edge creation. 

Given the equation for a ER network &lt;$k$&gt; = $p(N - 1)$, where $p = k / (N - 1)$. which gives us $p = 0.002$.


```python
#Average degree dist of Superhero

N_super = len(G_superhero.nodes())
k_super = G_superhero.number_of_edges()/N_super

p_random = k_super/(N_super-1) 
print(p_random)
```

```python
G1 = nx.erdos_renyi_graph(N_super, p_random)

out_ER = sorted([d for n, d in G1.degree()], reverse=True)

plt.hist([out_ER, out_super], bins=50, label=['ER out degree', 'Super out degree'])
plt.legend(loc='upper right')

plt.show()

#Sanity check
print('Total outgoing in ER:', sum(out_ER))
print('Total outgoing in Super Graph:', sum(out_super))
```

### Exercise 2: Let's build a simple visualization of network

```python
GC_super_un = GC_superhero.to_undirected()
```

```python
pos = nx.kamada_kawai_layout(GC_super_un)
```

```python
universe_color = []
for char in GC_super_un.nodes():
    if GC_super_un.nodes[char]['universe'] == 'dc':
        universe_color.append('#034078')
    elif GC_super_un.nodes[char]['universe'] == 'marvel':
        universe_color.append('#ff206e')
        
degree_list = []
for char in GC_super_un.nodes():
    degree_list.append(GC_super_un.degree(char))
    
#normalize node size to highest degree with max size = 30
max_degree = max(degree_list)
node_size_list = [ x/max_degree*200 for x in degree_list]
```

```python
plt.figure(figsize=(12,12)) 
nx.draw(GC_super_un, pos=pos, with_labels=False, node_size=node_size_list, width=0.05, node_color=universe_color, edge_color='#f9c80e')
```
