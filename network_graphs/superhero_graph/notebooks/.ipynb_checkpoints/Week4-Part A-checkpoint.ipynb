```python
import json
import urllib.request

import os.path


baseurl = "https://en.wikipedia.org/w/api.php?"
action = "action=query"
prop = "prop=revisions&rvprop=content" 
dataformat ="format=json"

char_file_list_dc = []
for character in dfdc['WikiLink']:
    #replace white space with _ for the query 
    character = character.replace(' ','_')
    character_url = urllib.parse.quote(character) #Use URL parser to convert special chars to URL friendly format
    char_file_list_dc.append(character_url+'.json')
    if os.path.isfile(f"{character_url}.json") == False: 
    #create query for each character
        title = f'titles={character_url}'
        query = "{}{}&{}&{}&{}".format(baseurl, action, prop, title, dataformat)
        #parse quote to format special characters 
        with urllib.request.urlopen(query) as wikiresponse:
            wikitext = wikiresponse.read().decode('utf-8')
        wikijson = json.loads(wikitext)
        #wiki_text_list.append(wikijson)
        with open(f"{character_url}.json", "w") as outfile:
            json.dump(wikijson, outfile)
            print(character_url)
    
    #else:
    #    print(character, " exists")
```

```python
dfdc_clean = dfdc.copy()

#We dont want to dismiss chars that likely link to their own page, 
#this means chars where some of the name is present in the link
#like char:Angel O'Day wikilink: Angel and the Ape 
#we do however, want to dismiss things like char: Yellow Lantern  wiki: Hal Jordan#Other versions 653

dim1 =dfdc_clean.shape

for character, wikilink, index in zip(dfdc_clean.CharacterName, dfdc_clean.WikiLink, dfdc_clean.index):
    #Turned out there was still majorty false positives,
    #if all(word not in wikilink for word in character.split(" ")):
        #dfdc_clean.drop(index, inplace=True)
    if character not in wikilink:
        #reviweing the cases that does not have an exact match
         dfdc_clean.drop(index, inplace=True)

dim2 = dfdc_clean.shape
print(f'A total of {(dim1[0]-dim2[0])} chars was deleted')
```

```python
#Check all the downloaded for redirects, the documents contain the exact phrase #REDIRECT

for index, file in dfdc_clean['file'].iteritems():
    f = open(file, 'r')
    redirect = re.findall(r'#redirect', f.read(), re.IGNORECASE)
    if redirect:
        print(index)
        print(f'Redirect found! in {file}, dropped row')
        dfdc_clean.drop(index, inplace=True)
        if os.path.exists(file) == True:
            print(file, 'is there')
            try: #Why Does this not work? THe file does exist but deleting it fails.
                os.remove(file)
                print(f'removed {file}')
            except:
                print("Error while deleting file ", file)
        
```

```python
import networkx as nx

G_dc = nx.Graph()
G_dc.clear()

for wikilink,file in zip(dfdc_clean.WikiLink, dfdc_clean.file):
    f = open(file, 'r')
    outlink = re.findall(r'\[\[([^:,.]+?)\]\]', f.read(), re.IGNORECASE)
    if outlink:
        for linkto in outlink:
            if linkto.split("|")[0] in dfdc_clean['WikiLink'].values:
                G_dc.add_edge(wikilink, linkto.split("|")[0])

print(sum(dict(G_dc.degree()).values()))
print(len(dict(G_dc.degree()).keys()))
```

Alternatively> Instead of adding edges between (and creating nodes) using the wikilink and 1st part of internal link, one could also create edges and nodes using the characternames as well as the 2nd part of the internal link.

Tried below:

```python
G2_dc = nx.Graph()
G2_dc.clear()

for char,file in zip(dfdc_clean.CharacterName, dfdc_clean.file):
    f = open(file, 'r')
    outlink = re.findall(r'\[\[([^:,.]+?)\]\]', f.read(), re.IGNORECASE)
    if outlink:
        for linkto in outlink:
            if linkto.split("|")[0] in dfdc_clean['WikiLink'].values:
                G2_dc.add_edge(char, linkto.split("|")[-1])
                
print(sum(dict(G2_dc.degree()).values()))
print(len(dict(G2_dc.degree()).keys()))
```

More links, and significantly more chars - how can that be?

Some of the character names & associated wiki links seem odd, the char name does not seem to match the url. This is for example the case for the char "Vectron" that has the wiki url "Elseworld's Finest: Supergirl & Batgirl#Secondary characters and cameos", and "Van'n Intraktus, Magaan" that has the wikilink "Green Lantern Corps#First attempt of rebuilding the Green Lantern Corps". This seem to arise from the situation when a wiki page contains a char links for certain character, that does not have their own wiki page, but instead is directed to another page that mention this char. Many of these incidens seem to arise from the page https://en.wikipedia.org/wiki/List_of_DC_Comics_characters:_V  where we have a list of links from character names, to wikilinks that are not specfically describing the character.

If we select what our csv believe is a character page, that in some cases may lead to a more general page, and we then download the content of this page and find wikilinks, and then create connections based on what we interpreted as a specific character based on that, we will end up creating a lot of connections to eg. 'Vectron', although the links was in fact just leading to the Green Lantern Corpse. 

As an example the above problem has linked the char Val Colby to the 'Lex Luthor#Relationships and family' wiki, and the consequence of the above is that the network think that Val Colby with the wikilink 'Lex Luthor#Relationships and family' has 32 connections, whereas in fact it has 0.

Problem is that the vissible wiki link (2nd part) is not nescesarily the name of the char, fx. "Doctor Fate|Fate" "Doctor Fate" and "Doctor Fate|Nabu" all appear as internal links, linking to the same wiki page. Hence the vissible link is sometimes showed as just Fate for short, and for another completely different reference, Nabu, links to the Doctor Fate page. For that reason it is better to stick to the WikiUrls that are unique. 

The most robust method is therefore to discard characters that does not seem to have their own specific page, before creating links from their wiki. Another method could be to create the character if mentioned, but not create any links based on the wiki-page if the wikipage does not seem to represent that specific character, but that would simply lead to isolated nodes as a character with no "legitimate" character wiki-page, would have no inbound or outbound leads.
